```python
import random as rn
```

```python
Study_hour = []
Department = []
age = []
Gender = []
IQ = []
Attendance = []
Sleep_hours = []
Internet_usage = []

for i in range (0, 100):
    Study_hour.append(rn.randint(20, 100))
    Department.append(rn.choice(["CSE", "BFT", "EEE", "JRN"]))
    age.append(rn.randint(20,26))
    Gender.append(rn.choice(["Male", "Female"]))
    IQ.append(rn.randint(80, 140))
    Attendance.append(rn.randint(0, 10))
    Sleep_hours.append(rn.randint(4, 9))
    Internet_usage.append(rn.randint(1, 10))
```

```python
St_dict = {
    "Study_hour": Study_hour, "Department": Department, "Age": age, "Gender": Gender,
    "IQ": IQ, "Attendance": Attendance, "Sleep_hours": Sleep_hours, "Internet_usage": Internet_usage
}
import pandas as pd

df = pd.DataFrame(St_dict)
df
```

```python
label = []
for i in range(100):
    if (Study_hour[i] > 50 and IQ[i] > 90 and Attendance[i] > 6):
        label.append(1)
    elif (Department[i] == "CSE" and Internet_usage[i] > 7 and Sleep_hours[i] < 5 and Study_hour[i]<30 and IQ[i] > 100):
        label.append(1)
    else:
        label.append(0)


df = pd.DataFrame(St_dict)
df["Outcome"] = label
df

```

```python
df.to_csv("synthetic_students.csv", index=False)
```
